## Introduction

A C-arm [fluoroscopy](https://en.wikipedia.org/wiki/Fluoroscopy) system is a specialized medical imaging device used primarily in surgical and interventional procedures. It consists of a C-shaped arm that houses an X-ray source on one end and a detector on the other. This configuration allows for real-time imaging of internal structures, organs, and tissues, aiding surgeons and interventional radiologists during procedures. Fluoroscopy is ubiquitously employed in [cardiovascular](https://www.jacc.org/action/showCitFormats?doi=10.1016/j.jacc.2004.10.014) [\[1\]](#Bibliography), [endovascular](https://link.springer.com/article/10.1007/s11548-018-1779-6) [\[2\]](#Bibliography), [orthopedic](https://journals.lww.com/jaaos/abstract/2019/10010/the_role_of_intraoperative_navigation_in.3.aspx) [\[3\]](#Bibliography), and [neurointerventions](https://link.springer.com/article/10.1007/s00586-011-2011-3) [\[4\]](#Bibliography), as well as for [pain management](https://link.springer.com/article/10.1007/s11916-018-0660-1) [\[5\]](#Bibliography) and [biopsies](https://pubs.rsna.org/doi/10.1148/radiology.220.1.r01jl29161) [\[6\]](#Bibliography) across diverse anatomical targets. To facilitate the positioning and orientation of the surgical instrument, the mobile C-arm must be positioned efficiently and accurately to obtain the desired radiographic view, leading to repeated image acquisitions known as [**fluoro-hunting**](https://www.spiedigitallibrary.org/journals/journal-of-medical-imaging/volume-5/issue-1/015005/Virtual-fluoroscopy-for-intraoperative-C-arm-positioning-and-radiation-dose/10.1117/1.JMI.5.1.015005.short) [\[7\]](#Bibliography). To assess the depth of a surgical instrument, the C-arm must be oriented at a 90-degree angle from the instrument’s trajectory and, after the depth is assessed, revert to the original pose to reassess the instrument trajectory. The necessity to reposition the C-arm and perform repeated image acquisition stems from the 2D projective nature of the X-ray radiograph. This contributes to the main limitations of fluoroscopy-guided interventions, namely the loss of depth perception and prolonged radiation exposure for both the [patient](https://www.sciencedirect.com/science/article/pii/S1051044309003443) [\[8\]](#Bibliography) and [medical staff](https://www.ajronline.org/doi/10.2214/AJR.16.16556) [\[9\]](#Bibliography).

Traditionally, medical residents learn how to operate and position the C-arm fluoroscopy by observing a more senior clinician during actual surgical interventions. This practice may lead to prolonged surgical times and excessive radiation exposure for both the patient and the medical staff.

## Objectives

The objective of this course project is to implement a miniaturized C-arm simulator to assist trainees (medical school students and medical residents) in understanding the geometrical relationship between the pose (position + orientation) of the C-arm and the acquired image.



## Background

[Simulation-based training for C-arm fluoroscopy positioning was previously shown to be effective for medical training](https://ir.lib.uwo.ca/etd/7231/) [\[10\]](#Bibliography). For this course project, we will utilize the software developed and published in [\[10\]](#Bibliography) and focus on hardware design and the integration of sensors and actuators.

The following video demonstrates how the software and the miniaturized C-arm simulator work.

[![IMAGE ALT TEXT](http://img.youtube.com/vi/HTMEmGuSuMk/0.jpg)](https://youtu.be/HTMEmGuSuMk?si=3_KXmIoWA2H2SW1f "C-arm Simulator")

Most clinical C-arm machines have 6 degrees-of-freedom (DOF).

<img src="https://drive.google.com/uc?id=1be7A-x0pwgVhvejOO9QkGIlPHgsDIpUV" alt="image" width="800" height="auto">

As shown in the video above, the *C* of the C-arm rotates about each of the sagittal, frontal, and longitudinal axes. The patient table translates along the longitudinal axis. Additionally, the base of the C-arm can be translated (it is often on wheels for smaller models) and rotates about the sagittal axis.


In this implementation, the miniaturized C-arm simulator utilizes **3** Bluetooth-enabled accelerometers/tilt sensors to measure the rotation of the C-arm, the rotation of the C-arm base, and the translation of the surgical table. The measurement data are transmitted to a PC running Windows 10/11 via wireless Bluetooth transmission.

<img src="https://drive.google.com/uc?id=1KxEmnyPMRJ6EUBcTeI5NcmmFLo0si63f" alt="image" width="800" height="auto">

The C-arm simulator shown in the figure above is roughly **10:1** scaled-down in size compared to a clinical C-arm fluoroscopy machine.

A [CT scan](https://www.nibib.nih.gov/science-education/science-topics/computed-tomography-ct) of a spine was obtained and virtually placed on the miniaturized surgical table through a process called [**registration**](https://www.mathworks.com/discovery/image-registration.html). The pose of the C-arm, representing the position and orientation of the X-ray transmitter and detector relative to the CT volume on the surgical table, is used to generate a synthetic X-ray image (i.e., simulated fluoroscopic image) through a process called [Digitally Reconstructed Radiograph, or DRR](https://www.cambridge.org/core/journals/journal-of-radiotherapy-in-practice/article/abs/role-of-digitally-reconstructed-radiographs-in-the-verification-process/32D533CAC6ABF67385BDF9FA2B5FEB65). In terms of mathematical principles, the DRR is generated using [volume rendering](https://en.wikipedia.org/wiki/Volume_rendering), commonly used in computer graphics and gaming engines.


## Instructions

For this course project, you are going to design a simplified version of the miniaturized C-arm simulator similar to the one shown in the figure above. The physical components of this simplified C-arm include:

1. The "C" of the C-arm, including a housing compartment for **Accelerometer 1** (refer to the figure above).
2. The surgical table, which comprises:
    - A flat surface that can be translated along the longitudinal direction.
    - A rotational knob that houses **Accelerometer 2**.
    - A compartment that houses a **geared motor** (i.e., from your previous course) such that, when activated, it translates the surgical table in a linear motion along the longitudinal direction.
3. A turntable (Turntable 1) with a sliding mechanism that houses the "C" of the C-arm.
4. The turntable itself is driven by another geared motor that rotates the C-arm about the frontal axis.
5. Turntable 2 and Accelerometer 3 **will not be implemented.**

### Software Companion

We will utilize the open-source software [3D Slicer](https://www.slicer.org/) to facilitate the transmission of the Bluetooth (BT) data and real-time animation. 3D Slicer is a platform commonly used in the medical imaging community, with extensions contributed by the research community available via the Extension Manager (similar to an App Store).

To use 3D Slicer, follow the instructions below:

1. [Download the latest version of 3D Slicer](https://download.slicer.org/). For this project, use **3D Slicer for Windows** and install it.
2. [Download the latest version of the PLUS Server](https://plustoolkit.github.io/download.html) and install it. (Note: The PLUS Server requires the use of a Windows PC.)
3. Pairing the Accelerometers with a Windows PC:
   - On your Windows PC, open Start Menu -> Settings -> Bluetooth and other Device Settings.
   - Turn on the Accelerometer (the blue light will start flashing).
     - The Accelerometer can be charged using a standard USB cable. Ensure it is charged before pairing.
     - **If using more than one accelerometer at a time**, turn on one accelerometer at a time.
   - Once the device is discovered, it should appear as "HC-06".
   - Enter **1234** as the PIN, then click Enter and wait for the pairing to complete.
4. Once paired, click on **More Bluetooth Options -> COM Port** and make a note of the outgoing COM Port number.
   - If working with multiple accelerometers, you may rename each BT accelerometer to make it easier to distinguish from others.
   - Repeat the pairing process for the second accelerometer.
5. Download [the configuration file **WMTest.txt**](https://owl.uwo.ca/x/QVwWZV), also available on OWL under Resources -> Project Materials.
   - Add the configuration in the PLUS Server Directory.
   - In the .txt file, change the **SerialPort** number to the outgoing COM Port number from Step 4.
   - After the modification, launch the PLUS Server.
6. Launch 3D Slicer. Install the following (you only need to do this once):
   - Open 3D Slicer -> View -> Extension Manager.
   - Install:
     - Slicer IGT, and
     - Slicer OpenIGTLink.
   - Restart 3D Slicer.
7. In 3D Slicer, under Modules -> drop-down menu -> IGT -> OpenIGTLinkIF, on the status checkbox, **select Active**. This assumes that the PLUS Server is concurrently running.
8. Download [the sample scene file **SampleScene.mrb**](https://owl.uwo.ca/x/nvjXTZ), also available directly on OWL.
   - Drag and drop it into 3D Slicer.
   - A 3D virtual model of a C-arm will appear. When moving/rotating the accelerometer, the C-arm will be animated.
   - You can also monitor the transform matrix changing in real-time as the accelerometer is moved.
9. If you are familiar with 3D Slicer, you will be able to extract the 3D model of the C-arm and save it as a 3D mesh file (i.e., .STL).

## Workflow

1. Based on the discussion in class, design 3D model parts of the C-arm simulator with the intention to be manufactured using a laser-cutter available in the Mechatronics Lab.
    - Confirm with the Lab Manager, Mr. Eugen Porter, about the materials available to you (e.g., acrylic sheet) and the capability of the laser cutter (maximum dimension).
    - Be mindful of the **thickness** of the parts: you may need to manufacture multiple identical parts and stack them to achieve the desired thickness.
2. In your design, incorporate the accelerometers. One accelerometer should be located at the C-arm tip (i.e., the location of the X-ray source), while the other should be located at the rotational axis of the patient table:
    - The first accelerometer is to measure the 3D orientation of the C-arm.
    - The second accelerometer is to measure the angular displacement (of the handle) to linear displacement (of the patient table).
3. Use the motor to actuate:
    - The C-arm, i.e., Turntable 1, with purely rotational displacement.
    - The patient table, i.e., Accelerometer 2, with rotation to linear displacement.
    - Carefully consider the torque and load required to move the C-arm and the patient table. Incorporate gears if necessary.
4. Accelerometers are used to generate the Digitally Reconstructed Radiograph (DRR) in real-time but can also be used to measure the response time of the actuators.
    - However, DRR generation is currently not implemented in the 3D Slicer sample scene.
    - Nonetheless, you will be able to animate the 3D models of the C-arm in 3D Slicer using the measurements from the BT accelerometers.
    - **To use accelerometer data to quantify the characteristics of the actuation system**, a *data synchronization* mechanism must be implemented.
        - As there is no easy way to synchronize the command sent to the motor via the micro-process with the data 3D Slicer receives from the accelerometers, you will not be required to implement such a synchronization mechanism.
        - However, you **must discuss how this may be accomplished**, i.e., how to interface data from multiple sources/systems asynchronously.
